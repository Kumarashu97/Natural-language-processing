```python
import pandas as pd
```

```python
message = pd.read_csv('spam.csv', encoding='latin-1')
```

```python

message = message.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
```

```python
message = message.rename(columns={'v1': 'label', 'v2': 'text'})
```

```python
message.head()
```

```python

##data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python

corpus = []
for i in range(0, len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['text'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python

##create the bag of words with n-grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True,ngram_range=(2,3)) #bigrams and trigrams
X = cv.fit_transform(corpus).toarray()
```
